```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as scipy
from scipy import optimize
import numdifftools as nd
def g_1(x):
    return 2*np.cos(x[0])**2+4*np.cos(x[1])
def g_2(x):
    return 1*np.sin(x[0])**3+3*np.sin(x[1])**2
x=np.linspace(0,np.pi,1000)
y=np.concatenate([np.linspace(0,np.pi/2,500),np.linspace(3*np.pi/2,2*np.pi,500)]) #ho una griglia 1000*1000
punti=[]
for i in range(0,len(x),1):
    for j in range(0,len(x),1):
        if abs(np.sin(x[i])-2*np.cos(y[j]))<0.001:   #condizione di Pareto critico CON TOLLERANZA
            punti.append([x[i],y[j]])
punti=np.array(punti)
l=len(punti)
print(l)
#ora faccio i conti del foglio con arcsin
punti2=np.concatenate([np.linspace(np.pi/3,np.pi/2,500),np.linspace(3*np.pi/2,5*np.pi/3,500)])
fig,axes=plt.subplots(1,2)
fig.set_size_inches(8,4)
axes[0].scatter([g_1(i) for i in punti],[g_2(i) for i in punti],s=0.8,color='blue')
axes[1].scatter([g_1([np.arcsin(2*np.cos(i)),i]) for i in punti2],[g_2([np.arcsin(2*np.cos(i)),i]) for i in punti2],s=0.8,color='red')
```
